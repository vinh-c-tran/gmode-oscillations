## NPE$\mu$ Big Apple
Steps
1. Construct the system of particles of interest by making a `baryon_list`, `meson_list`, and `lepton_list` as shown below. 
2. Read in the `.csv` output of the equilibrium structure calculation from the previous step. 
3. Call the `sound_speed_diff` function to calculate the 

```python
baryon_list = [Neutron, Proton]
meson_list = [sigma, omega, rho, phi] 
lepton_list = [electron, muon]
```

```python
npemu_bigapple = pd.read_csv('../equilibrium_structure/Example Output/npemu_bigapple.csv')
npemu_bigapple
```

```python
sound_npemu_bigapple = sound_speed_diff(baryon_list, meson_list, lepton_list, big_apple, npemu_bigapple, csv_name = './Output/bigapple_sound_speed_npemu')
sound_npemu_bigapple
```

### Plotting Results

```python
nb_array = sound_npemu_bigapple.loc[:, 'nB/n0']
cs2_ce2 = sound_npemu_bigapple.loc[:, 'cs^2 - ce^2']
plt.xlabel('nB/n0')
plt.ylabel('$c_s^2 - c_e^2$')
plt.title('NPE$\mu$ Big Apple')
plt.plot(nb_array, cs2_ce2)
plt.show()
```
