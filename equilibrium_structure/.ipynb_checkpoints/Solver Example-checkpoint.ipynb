```python
def plot_eos(dataframe, eos, colors = []):
    """ plot pressure vs energy density """
    
    nb_array = dataframe.loc[:, 'nB/n0']
    pressure = dataframe.loc[:, 'P (MeV/fm3)'] 
    energy_density = dataframe.loc[:, 'E Dens (MeV/fm3)']
    
    plt.plot(energy_density, pressure) 
    plt.xlabel('$\epsilon$ (MeV/fm3)')
    plt.ylabel('$P$ (MeV/fm3)')
    plt.title(eos.name + " EOS")
    plt.rcParams['xtick.top'] = True
    plt.rcParams['xtick.labeltop'] = False
    plt.rcParams['ytick.right'] = True
    
    plt.savefig(eos.name + ' eos.pdf')
```

```python
plot_eos(all_test, big_apple)
```

```python
def plot_eos_log(dataframe, eos, colors = []):
    """ plot pressure vs energy density """
    
    fig1, ax = plt.subplots()
    ax.set_box_aspect(1)
    nb_array = dataframe.loc[:, 'nB/n0']
    pressure = dataframe.loc[:, 'P (MeV/fm3)'] 
    energy_density = dataframe.loc[:, 'E Dens (MeV/fm3)']
    
    plt.plot(energy_density, pressure) 
    plt.xlabel('$\epsilon$ (MeV/fm3)')
    plt.ylabel('$P$ (MeV/fm3)')
    plt.title(eos.name + " EOS")
    plt.rcParams['xtick.top'] = True
    plt.rcParams['xtick.labeltop'] = False
    plt.rcParams['ytick.right'] = True
    
    plt.yscale('log')
    plt.xscale('log')
    
    plt.xlim(10**1, 10**3)
    plt.ylim(10**0, 10**3)
    
    plt.savefig(eos.name + ' eos_log_log.pdf')
```

```python
plot_eos_log(all_test, big_apple)
```

```python
def plot_fraction(dataframe, eos, colors = []):
    """ plot fractions for a single model """
    
    n0 = eos.n0
    nb_array = np.array(dataframe.loc[:, 'nB/n0'])
    hyp_bool = False 
    
    index = 0 
    line_style = ['-.', '--']
    
    for model in labels_dict:
        if model == eos: 
            label_position = labels_dict[model][0] 
            label_adjustment = labels_dict[model][1]
    
    legend_dict = {
        'Proton': '$p$',
        'Neutron': '$n$',
        'electron': '$e$', 
        'muon': '$\mu$', 
        'Lambda': '$\Lambda^0$',
        'Sigma_min': '$\Sigma^-$',
        'Sigma_neu': '$\Sigma^0$', 
        'Sigma_plus': '$\Sigma^+', 
        'Xi_min': '$\Xi^-$', 
        'Xi_neu': '$\Xi^0$'
    } 
    
    ax = plt.axes()
    for column_name in dataframe:
        if len(column_name.split()) > 1 and column_name.split()[1] == 'frac':
            ax.plot(nb_array * eos.n0, np.array(dataframe[column_name]),\
                     linestyle = line_style[index % 2], color = colors[index], label = legend_dict[column_name.split()[0]])
            plt.text(nb_array[label_position[index]] * eos.n0, np.array(dataframe[column_name][label_position[index]]) + label_adjustment[index], legend_dict[column_name.split()[0]])
            index += 1 
        if column_name.split()[0] == 'Lambda':
            hyp_bool = True 
    
    #plt.legend(bbox_to_anchor =(1.25, .95))
    #plt.title(eos.name + " Fraction")
    #plt.title('RMF-Gluck Fraction')
    plt.ylabel('$n_i/n_B$')
    plt.xlabel('$n_B$ (fm$^{-3}$)')
    plt.yscale('log')
    
    plt.ylim(10**-3,10**0)
    
    
    
    plt.rcParams['xtick.top'] = True
    plt.rcParams['xtick.labeltop'] = False
    plt.rcParams['ytick.right'] = True
    
    ax.set_aspect(0.43)
    
    if hyp_bool:
        file_name = eos.name + " frac_hyp.pdf"
    else: 
        file_name = eos.name + " frac.pdf"

    plt.savefig(file_name, bbox_inches = 'tight')
```

```python
def eos_plot_multiple(dataframe_dict, title_name, colors = []):
    """ plot multiple equation of states 
        dataframe_dict should be a dict such that the key are the dataframes
        and the values are the eos objects ie, [npe_big_apple: big_apple] 
    """
    index = 0 
    line_style = ['-.', '--', '-', ':']

    for model in dataframe_dict: 
        pressure = dataframe_dict[model].loc[:, 'P (MeV/fm3)'] 
        energy_density = dataframe_dict[model].loc[:, 'E Dens (MeV/fm3)']
        plt.plot(energy_density, pressure, linestyle = line_style[index % 4], color = colors[index], label = model.name)  
        index += 1 
        
    plt.xlabel('$\epsilon$ (MeV/fm3)')
    plt.ylabel('$P$ (MeV/fm3)')
    plt.legend()
    plt.title(title_name + " EOS")
    
    plt.set_aspect('equal', adjustable='box')
    
    plt.savefig(title_name + ' multiple_eos.pdf')
```

```python
eos_plot_multiple(data_dict_npe, 'NPE-$\mu$', flat_color_five)
```

```python
def eos_plot_multiple_new(dataframe, title_name, colors, matter_type = 'NPE-$\mu$'):
    """ eos plot multiple
        09_02_2022
    """
    
    #n0 = eos.n0 
    #nb_array = np.array(dataframe.loc[:, 'nB/n0']) 
    
    index = 0 
    line_style = ['-.', '--', '-', ':'] 
    hyp_bool = False 
    if matter_type != 'NPE-$\mu$':
        hyp_bool = True 
    

    ax = plt.axes()
    
    for model in dataframe:
        pressure = dataframe[model].loc[:, 'P (MeV/fm3)']
        energy_density = dataframe[model].loc[:, 'E Dens (MeV/fm3)'] 
        ax.plot(energy_density, pressure, linestyle = line_style[index % 4], color = colors[index], label = model.name)
        index += 1 
    
    plt.title(title_name + " EOS ")
    plt.ylabel('P (MeV/fm3)')
    plt.xlabel('$\epsilon$ (MeV/fm3)')
    #ax.set_aspect('equal', adjustable='box')
    #plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    plt.legend()
    
    #plt.ylim()
    
    plt.rcParams['xtick.top'] = True
    plt.rcParams['xtick.labeltop'] = False
    plt.rcParams['ytick.right'] = True
    
    #plt.xlim(0.08)
    
    #ax.tick_params(axis="y",direction="in", length = 4.5)
    #ax.tick_params(axis="x",direction="in", length = 4)
    #ax.xaxis.set_minor_locator(MultipleLocator(5))
    #ax.yaxis.set_minor_locator(MultipleLocator(5))
    ax.set_box_aspect(1)
    plt.xlim(0,1700)
    plt.ylim(0, 1000)
    #plt.yscale('log')
    
    #ax.set_aspect('equal')
    
    if hyp_bool:
        file_name = matter_type + " eos" + ".pdf"
    else:
        file_name = matter_type + " eos" + ".pdf"

    plt.savefig(file_name, bbox_inches='tight')
```

```python
def eos_plot_multiple_log(dataframe_dict, title_name, colors = []):
    """ plot multiple equation of states 
        dataframe_dict should be a dict such that the key are the dataframes
        and the values are the eos objects ie, [npe_big_apple: big_apple] 
    """
    index = 0 
    line_style = ['-.', '--', '-', ':']

    for model in dataframe_dict: 
        pressure = dataframe_dict[model].loc[:, 'P (MeV/fm3)'] 
        energy_density = dataframe_dict[model].loc[:, 'E Dens (MeV/fm3)']
        plt.plot(energy_density, pressure, linestyle = line_style[index % 4], color = colors[index], label = model.name)  
        index += 1 
        
    plt.xlabel('$\epsilon$ (MeV/fm3)')
    plt.ylabel('$P$ (MeV/fm3)')
    plt.legend()
    plt.title(title_name + " EOS")
    plt.xscale('log')
    plt.yscale('log')
    
    plt.savefig(title_name + ' multiple_eos.pdf')
```
